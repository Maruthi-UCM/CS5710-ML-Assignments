```python
import matplotlib.pyplot as plt

prog_lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]                             # intializing sample data
```

```python
# Creating pie chart
plt.pie(popularity, labels=prog_lang, autopct='%1.1f%%',wedgeprops={'linewidth': 1, 'edgecolor': 'black'})
plt.axis('equal')
plt.title('Popularity of Programming Languages')

# displaying pie chart
plt.show()
```
